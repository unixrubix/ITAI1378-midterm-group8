## Running detection with YOLO11 Large on video


```python
model = "YOLO11Large_video"
detector = YOLO("yolo11l.pt")
print(f"{model} loaded.")

detected_path = f"{base_path}/detected_{model}"
for filename in os.listdir(videos_path):
  if filename.endswith((".mp4")):
    file_path = os.path.join(videos_path, filename)
    print(f"Current file: {filename}")
    results = detector.predict(
        source=file_path,
        conf=0.12,           # Set confidence threshold lower
        classes=[14],        # 14 is the COCO index for 'bird'
        save=True,
        project=detected_path,
        name=filename,
        show=False,
        stream=True # Use stream=True to prevent memory accumulation for large videos
    )
    # Explicitly iterate through the results generator to process and free memory
    for r in results:
        pass # Do nothing with 'r' if not needed, just iterate to exhaust the generator
```
